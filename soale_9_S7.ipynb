```python
import numpy as np
import pandas as pd
import statsmodels.api as sm
import matplotlib.pyplot as plt
```

```python
data=pd.read_csv('soale 9 s7.csv')
```

```python
sm.qqplot(data['x'],line='s')
plt.show()
np.mean(data['x'])
```

```python
from scipy import stats
density=stats.gaussian_kde(data['x'])
xp=np.linspace(data['x'].min(),data['x'].max(),100)
yp=density.pdf(xp)
plt.plot(xp,yp,color='black')
plt.axvline(data['x'].mean(),color='r')
plt.show()
```
